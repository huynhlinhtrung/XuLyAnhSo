```python
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pylab as plt

# Read the image
data = iio.imread('fruit.jpg')
print(data.shape)

# Zoom the image by a factor of 2 (all axes)
bdata = nd.zoom(data, 2)
print(bdata.shape)

# Zoom the image by (2, 2, 1) for height, width, and channels
data2 = nd.zoom(data, (2, 2, 1))
print(data2.shape)

# Zoom the image by (0.5, 0.9, 1) for height, width, and channels
data3 = nd.zoom(data, (0.5, 0.9, 1))

# Show the result
plt.imshow(data3)
plt.show()
```

```python
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pyplot as plt

# Load the image named 'fruit.jpg'
# Ensure 'fruit.jpg' is in the same directory as this script, or provide the full path.
data = iio.imread('fruit.jpg')

# Print the shape of the loaded image data (height, width, channels)
print(data.shape)

# Rotate the image by 20 degrees, reshaping the output array to fit the entire rotated image
# This might introduce black borders to accommodate the new dimensions.
d1 = nd.rotate(data, 20)

# Display the first rotated image
plt.imshow(d1)

# Show the plot
plt.show()

# Rotate the image by 20 degrees, without reshaping the output array
# This means the output image will have the same dimensions as the original,
# and parts of the rotated image might be clipped.
d2 = nd.rotate(data, 20, reshape=False)

# Display the second rotated image
plt.imshow(d2)

# Show the plot
plt.show()
```

```python
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pyplot as plt

data = iio.imread('world_cup.jpg', mode="F")

print(data.shape)

V, H = data.shape

M = np.indices((V, H))

d = 5

q = 2 * d * np.random.rand(*M.shape) - d

mp = (M + q).astype(int)

d1 = nd.map_coordinates(data, mp)

plt.imshow(d1, cmap='gray')

plt.show()
```

```python
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pyplot as plt

# Hm bin i hnh hc
def GeoFun(outcoord):
    a = 10 * np.cos(outcoord[0] / 10.0) + outcoord[0]
    b = 10 * np.cos(outcoord[1] / 10.0) + outcoord[1]
    return a, b

# c nh mc xm dng float
data = iio.imread('world_cup.jpg', pilmode="F")

print(data.shape)

# p dng bin i hnh hc
d1 = nd.geometric_transform(data, GeoFun)

# Hin th nh kt qu
plt.imshow(d1, cmap='gray')
plt.title("nh sau khi p dng bin i hnh hc")
plt.axis('off')
plt.show()
```

```python
import numpy as np
import imageio.v2 as iio
import matplotlib.pyplot as plt
from scipy.ndimage import zoom

# c nh gc
data = iio.imread('fruit.jpg', pilmode='F')

# Phng to gp 3 ln
zoomed_up = zoom(data, 3.0)

# Thu nh 0.3 ln
zoomed_down = zoom(data, 0.3)

# Hin th nh gc, phng to, thu nh
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

axs[0].imshow(data, cmap='gray')
axs[0].set_title('nh gc')
axs[0].axis('off')

axs[1].imshow(zoomed_up, cmap='gray')
axs[1].set_title('Phng to x3')
axs[1].axis('off')

axs[2].imshow(zoomed_down, cmap='gray')
axs[2].set_title('Thu nh x0.3')
axs[2].axis('off')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import imageio.v2 as iio
import matplotlib.pyplot as plt
from scipy.ndimage import rotate

# c nh
data = iio.imread('fruit.jpg', pilmode='F')

# Xoay 45  vi reshape=True v reshape=False
rotated_true = rotate(data, 45, reshape=True)
rotated_false = rotate(data, 45, reshape=False)

# Hin th kt qu
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

axs[0].imshow(data, cmap='gray')
axs[0].set_title('nh gc')
axs[0].axis('off')

axs[1].imshow(rotated_true, cmap='gray')
axs[1].set_title('Xoay 45 (reshape=True)')
axs[1].axis('off')

axs[2].imshow(rotated_false, cmap='gray')
axs[2].set_title('Xoay 45 (reshape=False)')
axs[2].axis('off')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import imageio.v2 as iio
import matplotlib.pyplot as plt
import scipy.ndimage as nd

# c nh mc xm
data = iio.imread('world_cup.jpg', mode='L')
# Dilation v erosion vi size = (5, 5)
dilated = nd.grey_dilation(data, size=(5, 5))
eroded = nd.grey_erosion(data, size=(5, 5))

# Hin th kt qu
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

axs[0].imshow(data, cmap='gray')
axs[0].set_title('nh gc')
axs[0].axis('off')

axs[1].imshow(dilated, cmap='gray')
axs[1].set_title('Dilation (5,5)')
axs[1].axis('off')

axs[2].imshow(eroded, cmap='gray')
axs[2].set_title('Erosion (5,5)')
axs[2].axis('off')

plt.tight_layout()
plt.show()
```
